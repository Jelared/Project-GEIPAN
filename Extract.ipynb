```python
from bs4 import BeautifulSoup
import requests

villes_liste = []
phenos_liste = []
depart_liste = []
cas_liste = []
date_liste = []

for page_number in range(0,153):

    url = f"https://www.geipan.fr/fr/recherche/cas/tab?field_agregation_index_value=&field_latitude_value%5Bmin%5D=&field_latitude_value%5Bmax%5D=&field_longitude_value%5Bmin%5D=&field_longitude_value%5Bmax%5D=&field_departement_target_id&field_type_de_cas_target_id=All&field_phenomene_target_id&field_document_existe_ou_pas_value=All&field_is_new_value=All&field_is_revisited_value=All&field_date_d_observation_value%5Bmin%5D=&field_date_d_observation_value%5Bmax%5D=&field_date_value=&page={page_number}"
    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')
    villes = soup.find_all("td", class_="views-field views-field-title")
    phenos = soup.find_all("td", class_="views-field views-field-field-phenomene")
    departs = soup.find_all("td", class_="views-field views-field-field-departement-textuel")
    cass = soup.find_all("td", class_="views-field views-field-field-classification-des-cas")
    dates = soup.find_all("td", class_="views-field views-field-field-date-d-observation-textuel")

    for ville, pheno,depart,cas,date in zip(villes, phenos,departs,cass,dates):
        villes_liste.append(ville.text)
        phenos_liste.append(pheno.text)
        depart_liste.append(depart.text)
        cas_liste.append(cas.text)
        date_liste.append(date.text)
```

```python
import pandas as pd

df=pd.DataFrame({'Villes': villes_liste,'Dates': phenos_liste,'Départements': depart_liste,'cas':cas_liste, 'dates':date_liste})
df
```

```python
# Supprimer les derniers caractères de chaque ligne
df['Villes'] = df['Villes'].str[:-13]
df
```

```python
from tkinter import filedialog, Tk

# Création de la fenêtre Tkinter
root = Tk()
root.withdraw()  # Pour cacher la fenêtre principale

# Afficher la boîte de dialogue pour sélectionner l'emplacement de sauvegarde
file_path = filedialog.asksaveasfilename(defaultextension=".csv", filetypes=[("Fichiers CSV", "*.csv")])

if file_path:
    # Enregistrement du DataFrame dans un fichier CSV
    df.to_csv(file_path, index=False)
    print("Les données ont été sauvegardées dans", file_path)
else:
    print("Aucun emplacement de sauvegarde sélectionné.")
```
